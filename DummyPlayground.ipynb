```python
import torch
import pandas as pd
import numpy as np
from string import punctuation
from torch.utils.data import TensorDataset, DataLoader
from collections import Counter
```

```python
train_file = '../Train.csv'
val_file = '../Valid.csv'

traindf = pd.read_csv(train_file)
valdf = pd.read_csv(val_file)

traindf.head()
```

```python
print(traindf.loc[0])
```

```python
traindf['text'] = traindf['text'].apply(lambda x:x.lower())
valdf['text'] = valdf['text'].apply(lambda x:x.lower())
print(traindf['text'][0])
```

```python
print(punctuation)
```

```python
traindf['clean_text'] = traindf['text'].apply(lambda x:''.join([c for c in x if c not in punctuation]))
valdf['clean_text'] = valdf['text'].apply(lambda x:''.join([c for c in x if c not in punctuation]))
print(valdf['clean_text'][0])
```

```python
traindf['len_review'] = traindf['clean_text'].apply(lambda x:len(x))
valdf['len_review'] = valdf['clean_text'].apply(lambda x:len(x))
traindf.head()
```

```python
train_text = traindf['clean_text'].tolist()
val_text = valdf['clean_text'].tolist()

train_text = ' '.join(train_text)
val_text = ' '.join(val_text)

train_words = train_text.split()
val_words = val_text.split()

countwords_train = Counter(train_words)
countwords_val = Counter(val_words)
sorted_trainwords = countwords_train.most_common(len(countwords_train)) # returns list of top 'len(countwords_train)' elements
sorted_valwords = countwords_val.most_common(len(countwords_val)) # returns list of top 'len(countwords_val)' elements
print(sorted_valwords)
```

```python
x_train, y_train = torch.from_numpy(train_features), torch.from_numpy(train_labels)
x_val, y_val = torch.from_numpy(val_features), torch.from_numpy(val_labels)

print(f'shape of train data is {x_train.shape}')
print(f'shape of test data is {x_val.shape}')
print(f'shape of train labels is {y_train.shape}')
print(f'shape of test labels is {y_val.shape}')
```
